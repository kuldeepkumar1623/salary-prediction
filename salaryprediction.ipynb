```python
# Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
path = "/kaggle/input/positionsalaries1/Position_Salaries.csv"
df = pd.read_csv(path)

# Preview
df.head()
```

```python
X = df.iloc[:, 1:-1].values
y = df.iloc[:, -1].values
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)

pred = model.predict(X)
```

```python
plt.scatter(X, y)
plt.plot(X, pred)
plt.show()
```
